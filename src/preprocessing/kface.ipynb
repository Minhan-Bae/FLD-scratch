```python
from random import randint
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_lists = pd.read_csv("/data/komedi/backup/dataset/k_face_all_.csv").values.tolist()
```

```python
def visualize_landmark(image, landmarks):
    plt.figure(figsize=(11,11))
    plt.imshow(image)
    
    for idx, (x,y) in enumerate(landmarks):
        if idx >= 18:
            plt.scatter(x,y)
            plt.annotate(idx, (x,y))
```

```python
idx = randint(0,len(csv_lists)-1)

data = csv_lists[idx]

image = Image.open(data[3])

label = []
for landmark in data[8:]:
    x,y = eval(landmark[1:-1])
    label.append([x+data[4]-100,y+data[5]-100])
    
visualize_landmark(image,label)
```
